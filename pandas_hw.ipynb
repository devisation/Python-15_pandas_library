```python
def is_prime(num):
    if num == 2: return True
    else:
        for divisor in range(2, num):
            if num % divisor == 0: return False
        return True

def get_primes(length):
    primelist = []
    num = 2
    while(len(primelist) < length):
        if is_prime(num) and not num in primelist: primelist.append(num)
        num+=1
    return primelist

def get_fibnums(length):
    if length == 1: return [1]
    elif length == 2: return [1,1]
    elif length > 2:
        fiblist = [1,1]
        while(len(fiblist) < length):
            fiblist.append(fiblist[-1] + fiblist[-2])
        return fiblist
```

```python
df = pd.DataFrame({'Prime Numbers':get_primes(10), 
                   'Fibonacci Numbers':get_fibnums(10)},
                   index=['a','b','c','d','e','f','g','h','i','j'])
df
```

# Q11-14 (census and state_pop)

## 11

```python
state_pop_df = pd.read_csv(dflinks[3])
census_div_df = pd.read_csv(dflinks[4])
```

```python
state_pop = state_pop_df.copy()
census_div = census_div_df.copy()

census = census_div.sort_values('state').reset_index(drop=True)
states = state_pop.set_index('state')

for i,col in enumerate(states):
    census.insert(loc = i+3,
                  column = col,
                  value = census['state'].map(states.to_dict()[col]))
census
```

## 12

```python
censusT = census.set_index('state').transpose(copy=True)
census_yr = pd.DataFrame(censusT.loc['2010':].sum(axis=1))
census_yr.index.name = 'Year'
census_yr.columns = ['Population']
census_yr
```

## 13

```python
state_pops = censusT.loc['2010':]
census_summary = pd.DataFrame({'Total_Pop': state_pops.sum(axis=1),
                               'Avg_Pop': state_pops.mean(axis=1),
                               'Median_Pop': state_pops.median(axis=1)}, dtype=int)
census_summary
```

```python
df = pd.DataFrame({'Category': ksproj_df['main_category'],
                   'State': ksproj_df['state'], 
                   'pledged': ksproj_df['usd_pledged']}).dropna()
crdf = pd.crosstab(df.State, df.Category, 
                   df.pledged, 
                   aggfunc=lambda v: f'avg${np.mean(v):.2f}, med${np.median(v):.2f}, std${np.std(v):.2f}')
crdf
```

```python
df = pd.DataFrame({'Subcategory': ksproj_df['category'],
                   'State': ksproj_df['state'], 
                   'pledged': ksproj_df['usd_pledged']}).dropna()
crdf = pd.crosstab(df.State, df.Subcategory, 
                   df.pledged, 
                   aggfunc=lambda v: f'avg${np.mean(v):.2f}, med${np.median(v):.2f}, std${np.std(v):.2f}').fillna('N/A')
pd.set_option('Display.max_columns', None)
crdf
```
